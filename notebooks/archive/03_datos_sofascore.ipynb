```python
import requests
import pandas as pd
import json

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36",
    "Accept": "application/json"
}

TOURNAMENT_ID = 11653  # Liga de Primera — Chile

# Get available seasons for this tournament
response = requests.get(
    f"https://api.sofascore.com/api/v1/unique-tournament/{TOURNAMENT_ID}/seasons",
    headers=HEADERS
)
data = response.json()

for season in data["seasons"]:
    print(f"ID: {season['id']} — {season['name']} — Year: {season['year']}")

print(f"Status code : {response.status_code}")
print(f"Response    : {json.dumps(data, indent=2)[:500]}")
```
