```python
"""
Cell 11 â build_radar_light
Clean radar for light background (#e9e3d8).
No axis labels, no legend, no footer.
Only rings + exterior range numbers. % suffix only on aerialDuelsWonPercentage
and accuratePassesPercentage.
"""

THEME_LIGHT = {
    "bg":     "#e9e3d8",
    "ring":   "#ccc5b8",
    "edge":   "#9a9080",
    "p1":     "#0047CC",
    "p2":     "#E8000D",
    "avg":    "#3A3530",
    "label":  "#1A1A1A",
    "range":  "#2E2A26",
    "credit": "#9E9790",
}

# Metrics that use a % suffix in the range labels
PCT_METRICS = {"aerialDuelsWonPercentage", "accuratePassesPercentage"}

def build_radar_light(player_name: str, position: str) -> None:

    p1 = get_player(df_2025, player_name)
    p2 = get_player(df_2026, player_name)
    if p1 is None or p2 is None:
        return

    metrics = RADAR_METRICS[position]
    pool    = _get_pool(position)
    low, high, median = _get_scale(pool, metrics)

    vals1 = [_fmt(p1[m], m) for m in metrics]
    vals2 = [_fmt(p2[m], m) for m in metrics]
    _invert(metrics, low, high, vals1, vals2, median)

    T = THEME_LIGHT

    # Identify which radar axis positions correspond to % metrics
    pct_positions = {i for i, m in enumerate(metrics) if m in PCT_METRICS}

    labels = [""] * len(metrics)

    radar = Radar(
        labels, low, high,
        num_rings=4, ring_width=1, center_circle_radius=1,
    )
    fig, ax = radar.setup_axis()
    fig.set_facecolor(T["bg"])
    ax.set_facecolor(T["bg"])
    fig.set_size_inches(9, 9, forward=True)
    fig.set_dpi(200)

    radar.draw_circles(
        ax=ax,
        facecolor=T["ring"], edgecolor=T["edge"], lw=1.5, alpha=0.7,
    )

    radar.draw_range_labels(
        ax=ax, fontsize=10, color=T["range"], fontweight="bold",
    )

    # Append % to range labels only for percentage-based metrics.
    # mplsoccer draws num_rings labels per axis, so we group by num_rings
    # and patch only the axes in pct_positions.
    num_rings = 4
    all_texts = [t for t in ax.texts]
    for idx, txt in enumerate(all_texts):
        axis_idx = idx // num_rings   # which axis this label belongs to
        if axis_idx in pct_positions:
            try:
                val = float(txt.get_text())
                txt.set_text(f"{int(round(val))}%")
                txt.set_fontweight("bold")
                txt.set_fontsize(10)
            except (ValueError, TypeError):
                pass

    # League average â heavily dashed
    radar.draw_radar_solid(median, ax=ax, kwargs={
        "facecolor": "none",
        "edgecolor": T["avg"],
        "linewidth": 3.2,
        "linestyle": (0, (5, 2)),
        "alpha": 1.0,
        "zorder": 2,
    })

    # 2025 â electric blue
    radar.draw_radar_solid(vals1, ax=ax, kwargs={
        "facecolor": T["p1"],
        "alpha": 0.40,
        "edgecolor": T["p1"],
        "linewidth": 3.8,
        "zorder": 3,
    })

    # 2026 â pure red
    radar.draw_radar_solid(vals2, ax=ax, kwargs={
        "facecolor": T["p2"],
        "alpha": 0.45,
        "edgecolor": T["p2"],
        "linewidth": 4.2,
        "zorder": 4,
    })

    plt.tight_layout(pad=0.5)

    base = f"{player_name.replace(' ', '_')}_2025_vs_2026_radar_light"
    _save(fig, f"{base}.png")
    _save(fig, f"{base}_transparent.png", transparent=True)
    plt.show()
```
